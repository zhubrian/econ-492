```python
import datetime
import numpy as np
import pandas as pd
import time
from datetime import date, timedelta
from pycoingecko import CoinGeckoAPI
cg = CoinGeckoAPI()
```

# Get coin list

```python
def get_date_list(start_date,stop_date):
    date_list = []
    curr_date = start_date
    while curr_date <= stop_date:
        date_list.append(curr_date.strftime("%d-%m-%Y"))
        curr_date += timedelta(days=1)
    return date_list
```

```python
coin_list_full = []
pages = 6
per_page = 100
for p in range(pages):
    coin_list_page = cg.get_coins_markets(vs_currency='USD',page=p)
    for pp in range(per_page):
        coin_list_full.append(coin_list_page[pp]['id'])
coin_list_full = coin_list_full[100:]
```

```python
start_date = date(2020,3,25)
stop_date = date(2022,3,25)
date_list = get_date_list(start_date,stop_date)
len(date_list)
```

```python
coin_list = []
for i in range(len(coin_list_full)):
    hist_beg = cg.get_coin_history_by_id(date=date_list[0],id=coin_list_full[i],localization='false')
    time.sleep(1.201)
    hist_end = cg.get_coin_history_by_id(date=date_list[-1],id=coin_list_full[i],localization='false')
    try:
        hist_data = np.array([hist_beg['market_data']['current_price']['usd'],
                              hist_beg['market_data']['market_cap']['usd'],
                              hist_beg['market_data']['total_volume']['usd'],
                              float(hist_beg['community_data']['reddit_accounts_active_48h']),
                              hist_end['market_data']['current_price']['usd'],
                              hist_end['market_data']['market_cap']['usd'],
                              hist_end['market_data']['total_volume']['usd'],
                              float(hist_end['community_data']['reddit_accounts_active_48h'])])
        if (float(hist_beg['community_data']['reddit_accounts_active_48h']) > 0) and (float(hist_end['community_data']['reddit_accounts_active_48h']) > 0):
            coin_list.append(coin_list_full[i])
            print("accepted",coin_list_full[i],i)
        else:
            print("rejected",coin_list_full[i],i)
        time.sleep(1.201)
    except:
        print("rejected",coin_list_full[i],i)
        time.sleep(1.201)
        
```

# Get coin data

```python
start_date = date(2020,3,25)
stop_date = date(2022,3,24)
date_list_final = get_date_list(start_date,stop_date)
coin_list_final = coin_list[:100]

print("coins:",len(date_list_final))
print("dates:",len(coin_list_final))
print("hours:",len(date_list_final)*len(coin_list_final)/3000)
```

```python
price = np.zeros((len(date_list_final),len(coin_list_final)))
mktcap = np.zeros((len(date_list_final),len(coin_list_final)))
volume = np.zeros((len(date_list_final),len(coin_list_final)))
social = np.zeros((len(date_list_final),len(coin_list_final)))
for t in range(len(date_list_final)):
    for i in range(len(coin_list_final)):
        hist = cg.get_coin_history_by_id(date=date_list_final[t],id=coin_list_final[i],localization='false')
        try:
            price[t,i] = hist['market_data']['current_price']['usd']
            mktcap[t,i] = hist['market_data']['market_cap']['usd']
            volume[t,i] = hist['market_data']['total_volume']['usd'] 
            social[t,i] = float(hist['community_data']['reddit_accounts_active_48h'])
        except KeyError:
            price[t,i] = price[t-1,i]
            mktcap[t,i] = mktcap[t-1,i]
            volume[t,i] = volume[t-1,i]
            social[t,i] = social[t-1,i]
            print(date_list_final[t],coin_list_final[i],"EXCEPTION")
        time.sleep(1.201)
    print(date_list_final[t],"done")
```

```python
price_df = pd.DataFrame(price,index=date_list_final,columns=coin_list_final)
mktcap_df = pd.DataFrame(mktcap,index=date_list_final,columns=coin_list_final)
volume_df = pd.DataFrame(volume,index=date_list_final,columns=coin_list_final)
social_df = pd.DataFrame(social,index=date_list_final,columns=coin_list_final)

price_df.to_csv(r"C:\Users\zhubr\OneDrive\Desktop\ECON 492\data\prices.csv")
mktcap_df.to_csv(r"C:\Users\zhubr\OneDrive\Desktop\ECON 492\data\mktcap.csv")
volume_df.to_csv(r"C:\Users\zhubr\OneDrive\Desktop\ECON 492\data\volume.csv")
social_df.to_csv(r"C:\Users\zhubr\OneDrive\Desktop\ECON 492\data\social.csv")
```

# Get T-bill data

```python
tbill1m = pd.read_csv(r"C:\Users\zhubr\OneDrive\Desktop\ECON 492\data\tbill1m.csv")
tbill1m_date_list = []
for t in range(len(tbill1m)):
    curr_date = datetime.datetime.strptime(tbill1m['Date'][t],'%d/%m/%Y')
    tbill1m_date_list.append(curr_date.strftime("%d-%m-%Y"))
```

```python
rfrate = np.zeros(len(date_list_final))
for s in range(len(date_list_final)):
    date_found = 0
    for t in range(len(tbill1m_date_list)):
        if date_list_final[s] == tbill1m_date_list[t]:
            rfrate[s] = tbill1m['1M'][t]
            date_found = 1
    if date_found == 0:
        rfrate[s] = rfrate[s-1]
```

```python
rfrate_df = pd.DataFrame(rfrate,index=date_list_final,columns=['tbill_1m'])
rfrate_df.to_csv(r"C:\Users\zhubr\OneDrive\Desktop\ECON 492\data\rfrate.csv")
```
